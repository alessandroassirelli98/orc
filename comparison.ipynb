```python
guesses = np.load('/tmp/mu0.npy',allow_pickle=True).item()
U0 = guesses['U']
KK0 = guesses['KK']

guesses = np.load('/tmp/mu10.npy',allow_pickle=True).item()
U10 = guesses['U']
KK10 = guesses['KK']

N = U10.shape[0]
```

```python
z_err1 = np.ones((N, 4))
KK0_control = np.array([KK0[i] @ z_err1[i] for i in range(N)])
KK10_control = np.array([KK10[i] @ z_err1[i] for i in range(N)])
time_vec = np.linspace(0.0,N*0.005,N)

KK0_norm = np.array([np.linalg.norm(KK0[i])**2 for i in range(N)])
KK10_norm = np.array([np.linalg.norm(KK10[i])**2 for i in range(N)])
```

```python
plt.figure()
plt.title("2-norm of K")
# plt.plot(time_vec, KK0_control[:,0])
# plt.plot(time_vec, KK10_control[:,0])
plt.plot(time_vec, KK0_norm)
plt.plot(time_vec, KK10_norm)
plt.axvline(int(N/8)*0.005, linestyle="dashed", color="black")
plt.axvline(int(N/4)*0.005, linestyle="dashed", color="black")
plt.axvline(int(N/2)*0.005, linestyle="dashed", color="black")
plt.xlabel("t [s]")
plt.legend(["mu_factor 0", "mu_factor 10"])

plt.subplot(1,3,2)
plt.plot(time_vec, U0[:,0])
plt.plot(time_vec, U10[:,0])
plt.axvline(int(N/8)*0.005, linestyle="dashed", color="black")
plt.axvline(int(N/4)*0.005, linestyle="dashed", color="black")
plt.axvline(int(N/2)*0.005, linestyle="dashed", color="black")
plt.legend(["mu_factor 0", "mu_factor 10"])

plt.subplot(1,3,3)
plt.plot(time_vec, U0[:, 0] + KK0_control[:,0])
plt.plot(time_vec, U10[:, 0] + KK10_control[:,0])
plt.axvline(int(N/8)*0.005, linestyle="dashed", color="black")
plt.axvline(int(N/4)*0.005, linestyle="dashed", color="black")
plt.axvline(int(N/2)*0.005, linestyle="dashed", color="black")
plt.legend(["mu_factor 0", "mu_factor 10"])

```
